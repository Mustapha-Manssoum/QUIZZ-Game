### Using the microphone as audio input

pour une détection efficace il faut prononcer les phrases suivantes : 
* Le un 
* Le deux
* le trois
* Le quatre
* Oui
* Non

```python
recognizer = speech_recognition.Recognizer()
x=1
while(x):
    try:
        with speech_recognition.Microphone() as mic:
            recognizer.adjust_for_ambient_noise(mic, duration=0.2)
            audio = recognizer.listen(mic)
            print(type(audio))
            text = recognizer.recognize_google(audio, language='fr-FR' )
            text = text.lower()
            if len(text) > 0:
                x=0
                print(f'Recognized : {text}')
                break
    
    except speech_recognition.UnknownValueError():
        recognizer = speech_recognition.Recognizer()
        continue
```

```python
# La variable text stocke la réponse prononcé par l'utilisateur
# Cette variable doit être comparée avec les réponses que nous avons définie pour notre quizz dans une base de données.
print(text)
```

## Using an audio file as input to the model
Une autre manière de faire c'est d'utiliser une entrée audio autre que celle du microphone

### <span style="color:green;">Load the audio file and read it</span>

```python
import os
import librosa
import IPython.display as ipd

audio_path = os.getcwd() + '\\data\\quatre\\a3865408-17ce-4e98-b8ad-c258d83001ce.mp3'
data, sr = librosa.load(audio_path)
ipd.Audio(data, rate=sr)
```

### <span style="color:green;">Save the audio as a WAV file</span>

```python
import soundfile as sf
sf.write("example.wav", data, sr)
ipd.Audio("example.wav")
```

### <span style="color:green;">Speech recognition process</span>

```python
recognizer = speech_recognition.Recognizer()

with speech_recognition.AudioFile("example.wav") as wavaudio:
    audio = recognizer.record(wavaudio, duration=1)
    
    # recognize speech using Google Speech Recognition
    try:
        text = recognizer.recognize_google(audio, language='fr-FR')
        text = text.lower()
    except speech_recognition.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except speech_recognition.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))

    print(text)
```
